Este notebook demonstra o uso do agente para análise de dados de e-commerce em linguagem natural.

## 1. Configuração

```python
import os
from dotenv import load_dotenv

load_dotenv()  # carrega GEMINI_API_KEY e DB_PATH do arquivo .env

# Se preferir, defina diretamente (nao commitar com chave real):
# os.environ["GEMINI_API_KEY"] = "sua_chave_aqui"

os.environ.setdefault("DB_PATH", "banco.db")

from agent import EcommerceAgent

agent = EcommerceAgent()
print("✅ Agente inicializado com sucesso!")
```

## 2. Análise de Vendas e Receita

```python
resposta = agent.ask('Quais são os 10 produtos mais vendidos?')
print(resposta)
```

```python
resposta = agent.ask('Qual é a receita total por categoria de produto?')
print(resposta)
```

## 3. Análise de Entrega e Logística

```python
resposta = agent.ask('Qual é a quantidade de pedidos por status?')
print(resposta)
```

```python
resposta = agent.ask('Qual é o percentual de pedidos entregues no prazo por estado dos consumidores?')
print(resposta)
```

## 4. Análise de Satisfação e Avaliações

```python
resposta = agent.ask('Qual é a média de avaliação geral dos pedidos?')
print(resposta)
```

```python
resposta = agent.ask('Quais são os top 10 vendedores com melhor média de avaliação?')
print(resposta)
```

## 5. Análise de Consumidores

```python
resposta = agent.ask('Quais estados têm maior volume de pedidos e maior ticket médio?')
print(resposta)
```

```python
resposta = agent.ask('Quais estados têm maior atraso médio nas entregas?')
print(resposta)
```

## 6. Análise de Vendedores e Produtos

```python
resposta = agent.ask('Quais produtos são mais vendidos em cada estado?')
print(resposta)
```

```python
resposta = agent.ask('Quais categorias têm maior taxa de avaliação negativa (nota 1 ou 2)?')
print(resposta)
```
